```python
import math
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

# 威力導演

```python
lo = np.array([[267,304],[286,463]])  # x1,x2, y1,y2    (267,286):端點 - (304,463):中心點 = 方向
lo = lo*1.333333333333333
lo
```

```python
def angle_of_vector(x, y):
    angle = math.degrees(math.atan2(-y, x))   # y軸是相反的
    if angle<0:
        return angle+360
    return angle
     

def angle_of_line(x1, y1, x2, y2): # 端點 中心
    angle = math.degrees(math.atan2(-(y1-y2), x1-x2))
    if angle<0:
        return angle+360
    return angle 
```

```python
angle_of_line(lo[0][0], lo[1][0], lo[0][1], lo[1][1])
```

```python
img    = cv.imread('../dataset/temp_img/8_15.png')
RGBimg = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.figure(figsize = (10,5))
plt.scatter(lo[0],lo[1])
plt.plot(lo[0], lo[1], color="yellow", linewidth=5)

plt.imshow(RGBimg)
plt.title('dirction')
plt.show()
```
