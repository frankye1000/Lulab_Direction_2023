```python
# create a CLAHE object
def clahe(img):
    clahe_ = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    cl1    = clahe_.apply(img)
    return cl1


def equ(img):
    equ_ = cv.equalizeHist(img)
    return equ
```

```python
img   = cv.imread('../dataset/test_img/000140.png',0)
equ   = cv.equalizeHist(img)
# clahe = clahe(img)
blur  = cv.GaussianBlur(equ, (19, 19), 0)
edges = cv.Canny(blur, 80, 100)

plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.show()
plt.imshow(equ,cmap='gray')
plt.title('equalizeHist Image')
plt.show()

plt.imshow(blur,cmap='gray')
plt.title('blur Image')
plt.show()

plt.imshow(edges,cmap='gray')
plt.title('Edge Image')
plt.show()
```
